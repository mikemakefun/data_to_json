```python
from csv import DictReader

with open("wine-ratings.csv", encoding="utf-8", errors="ignore") as temp_file:
    reader = DictReader(temp_file)
    wines = list(reader)
```

Sorting wine via following creteria:

- Type of wine (red or white)
- Wine score range or baseline (e.g. between 90 and 92 points)
- Region

```python
import json
sorted_wine = []
for wine in wines:
    if "Red" or "White" in wine["variety"]:
        if float(wine["rating"]) >= 90.0 and float(wine["rating"]) <= 92.0:
            sorted_wine.append(wine)

with open("csv_saved.json","w") as opened_file:
    json.dump(sorted_wine,opened_file)
```
